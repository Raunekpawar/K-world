```python
import numpy as np
import pandas as pd
kdrama = pd.read_csv('top100_kdrama.csv')
kdrama.head()
```

```python
kdrama.info()
```

```python
kdrama = kdrama[['Name','Aired Date','Number of Episode','Network','Content Rating','Synopsis','Cast','Genre','Tags','Rank','Rating']]
```

```python
kdrama.head()
```

```python
kdrama['Synopsis'][0]
        
```

```python
kdrama['Tags'][0]
```

```python
kdrama['tags'] = kdrama['Synopsis'] + kdrama['Cast'] + kdrama['Genre'] + kdrama['Tags'] + kdrama['Content Rating']
```

```python
kdrama.head()
```

```python
new_df = kdrama[['Rank','Name','Aired Date','Number of Episode','Network','tags']]
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(kdrama):
    kdrama_index = new_df[new_df['Name'] == kdrama].index[0]
    distances = similarity[kdrama_index]
    kdrama_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in kdrama_list:
        print(new_df.iloc[i[0]].Name)
```

```python
recommend('My Mister')
```
